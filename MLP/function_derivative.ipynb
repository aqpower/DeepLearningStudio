```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import subprocess
import os

# 检查并创建 activation_func_images 文件夹
if not os.path.exists("activation_func_images"):
    os.makedirs("activation_func_images")

def plot_derivative_sympy(expr, name):
    """
    对给定的 SymPy 表达式进行求导并绘制图像。

    参数：
    expr (SymPy expression): 待求导的 SymPy 表达式
    name (str): 函数的名称，用于图像标题

    返回：
    无
    """
    # 定义变量
    x = sp.symbols("x")
    x_vals = np.linspace(-5, 5, 200)

    # 计算函数的导数
    derivative_expr = sp.diff(expr, x)

    y_vals = [expr.subs(x, i).evalf() for i in x_vals]
    y_prime_vals = [derivative_expr.subs(x, i).evalf() for i in x_vals]

    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, y_vals, label=name)
    plt.plot(x_vals, y_prime_vals, label=f"{name} Prime")
    plt.legend()
    plt.title(f"Function and Derivative of {name}")
    plt.xlabel("x")
    plt.ylabel("y")
    # plt.grid(True)
    # 定义 SVG 文件和 EMF 文件的路径
    svg_path = f"./activation_func_images/{name}_plot.svg"
    emf_path = f"./activation_func_images/{name}_plot.emf"
    plt.savefig(svg_path)

    # 使用 Inkscape 将 SVG 转换为 EMF
    subprocess.run(f"inkscape --export-filename={emf_path} {svg_path}", shell=True)
    plt.show()
```

```python
x = sp.symbols("x")
```

$$\text{sigmoid}(x) = \frac{1}{1 + e^{-x}}$$


```python
sigmoid = 1 / (1 + sp.exp(-x))
```

$$\text{tanh}(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

```python
tanh = sp.tanh(x)
```

$$\text{ReLU}(x) = \begin{cases} 0 & \text{if } x < 0 \\ x & \text{if } x \geq 0 \end{cases}$$

```python
relu = sp.Max(0, x)
```

$$\text{Leaky ReLU}(x) = \begin{cases} 0.01x & \text{if } x < 0 \\ x & \text{if } x \geq 0 \end{cases}$$

```python
leaky_relu = sp.Max(0.01 * x, x)
```

$$\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{N} e^{x_j}}$$

$$
\text{softmax}(x_1, x_2, \ldots, x_n) = \left( \frac{e^{x_1}}{e^{x_1} + e^{x_2} + \ldots + e^{x_n}}, \frac{e^{x_2}}{e^{x_1} + e^{x_2} + \ldots + e^{x_n}}, \ldots, \frac{e^{x_n}}{e^{x_1} + e^{x_2} + \ldots + e^{x_n}} \right)
$$


当 $N = 2$ 时，$Softmax$ 函数的表达式可以简化为：

$$
\text{softmax}(x_1, x_2) = \left( \frac{e^{x_1}}{e^{x_1} + e^{x_2}}, \frac{e^{x_2}}{e^{x_1} + e^{x_2}} \right)
$$

当 $N = 1$ 时，$Softmax$ 函数的表达式可以简化为：

$$
\text{softmax}(x) = \frac{e^{x}}{e^{x}} = 1
$$

为便于表示，这里采用 $N = 1$ 时的 $softmax$ 表达式。


```python
softmax = sp.Integer(1)
```

```python
# 绘制图像
plot_derivative_sympy(sigmoid, "Sigmoid")
plot_derivative_sympy(tanh, "tanh")
plot_derivative_sympy(relu, "ReLU")
plot_derivative_sympy(leaky_relu, "Leaky_ReLU")
plot_derivative_sympy(softmax, "SoftMax")
```
